# ED implementation in Fixed $M_z$ and $M_x$ subspace.

**Heisenberg Hamiltonian (for chain)**: 


$$\mathcal{H}=  J\sum\limits_{i}^{N-1}\hat{S}_{i}\cdot\hat{S}_{i+1} = J\sum\limits_{i}^{N-1}H_{i,i+1}$$
We can also re-write the Hamiltonian as:
$$\mathcal{H}=J\sum_{i=1}^{N-1}
\hat{S}^z_i \hat{S}^z_{i+1} +
\frac{1}{2}\left[
\hat{S}^+_i \hat{S}^-_{i+1} +
\hat{S}^-_i \hat{S}^+_{i+1}
\right] = H_{d} + H_{od}
$$

States representation::
--------------
Computational basis is  the basis of
$N$-spins that can be represented by using the $S^z$ projection
($\uparrow,\downarrow$) of each spin as: $|s_1,s_2,..., s_N\rangle$ = $|s_1\rangle \otimes |s_2\rangle \otimes \cdots \otimes |s_N\rangle$.

For example in the computer language, a basis state is simply given by an integer, e.g. the number `4` has the following sequence of digits $\quad$0100 (first 4  bits from left for L = 4 and so on) 


$$\begin{eqnarray}
\mid \downarrow \downarrow \uparrow \uparrow \rangle & \equiv &|0011\rangle
\equiv |3\rangle , \\
\mid \downarrow \uparrow \downarrow \uparrow \rangle &\equiv &|0101\rangle
\equiv |5\rangle , \\
\mid \downarrow \uparrow \uparrow \downarrow \rangle &\equiv &|0110\rangle
\equiv |6\rangle , \\
\mid \uparrow \downarrow \downarrow \uparrow \rangle &\equiv &|1001\rangle
\equiv |9\rangle , \\
\mid \uparrow \downarrow \uparrow \downarrow \rangle &\equiv &|1010\rangle
\equiv |10\rangle , \\
\mid \uparrow \uparrow \downarrow \downarrow \rangle &\equiv &|1010\rangle
\equiv |12\rangle .
\end{eqnarray}
$$

Example of Hamiltonian matrix elements filling: $H_d[3][3] = \langle  \downarrow \downarrow \uparrow \uparrow \mid H_{d} \mid \downarrow \downarrow \uparrow \uparrow \rangle = J/2$


Demonstrating above part with an Example (4-spin problem)
-------------------
In the spin-1/2
Heisenberg model, a matrix with $2^{N}\times 2^{N}$ elements can be
reduced to $2S+1$ smaller matrices, corresponding to the projections $%
m=(N_{\uparrow }-N_{\downarrow })/2=-S,-S+1,...,S-1,S$, with $S=N/2$.
For example, in a 4-site problem the full Hilbert-space dimension is 16 ($2^4$) while this can be reduced to 6 ($^{4}C_2$ combinations, $\because$ $N_{\uparrow}=2$) in a fixed magnetization sector of $M_z=0$. The 6 basis states that contribute to he Hamiltonian matrix can then be obtained as:

<span>After applying the Hamiltonian on our basis states, the
reader can verify as an exercise that we obtain </span>
$$\begin{eqnarray}
{{\hat{H}}}\,|0101\rangle &=&-J\,|0101\rangle +\frac{J}{2}\left[
\,|1100\rangle +\,|1001\rangle +\,|0011\rangle +\,|0110\rangle \right] , \\
{{\hat{H}}}\,|1010\rangle &=&-J\,|1010\rangle +\frac{J}{2}\left[
\,|1100\rangle +\,|1001\rangle +\,|0011\rangle +\,|0110\rangle \right] , \\
{{\hat{H}}}\,|0011\rangle &=&\frac{J}{2}\left[ \,\,|0101\rangle
+\,|1010\rangle \right] , \\
{{\hat{H}}}\,|0110\rangle &=&\frac{J}{2}\left[ \,\,|0101\rangle
+\,|1010\rangle \right] , \\
{{\hat{H}}}\,|1001\rangle &=&\frac{J}{2}\left[ \,|0101\rangle +\,|1010\rangle
\right] , \\
{{\hat{H}}}\,|1100\rangle &=&\frac{J}{2}\left[ \,|0101\rangle +\,|1010\rangle
\right] .%
\end{eqnarray}$$

<span>The resulting matrix for the operator </span>$\hat{H}$<span> in the
</span>$S^{z}$<span> representation is: </span> $$H=J\left(
\begin{array}{llllll}
0 & 1/2 & 0 & 0 & 1/2 & 0 \\
1/2 & -1 & 1/2 & 1/2 & 0 & 1/2 \\
0 & 1/2 & 0 & 0 & 1/2 & 0 \\
0 & 1/2 & 0 & 0 & 1/2 & 0 \\
1/2 & 0 & 1/2 & 1/2 & -1 & 1/2 \\
0 & 1/2 & 0 & 0 & 1/2 & 0
\end{array}
\right) .  $$

**Exact diagonalization is all about constructing the above matrix and diagonalizing it.**

**Importance of Symmetries:**
------
Motivation in getting the information of conserved quantities or symmetric operations is to make a block-diagonal form of Hamiltonian where each subspace becomes independent thereby reducing the Hilbert space dimension drastically.

(1) Consider the magnetization operator, $\mathcal{S}^z_T = \frac{1}{N}\sum\limits_i \hat{S}^z_i$.

We can see that,

\begin{eqnarray}
[\mathcal{S}^z_T,H] &=& 0, \quad \textbf{Magnetization is a conserved quantity}\nonumber \\
\therefore \mathcal{S}^z_T|\psi\rangle &=& M_z|\psi\rangle, \quad M_z \in \{0, \pm 1, \pm 2, ... \}
\end{eqnarray}



(2) Similarly,  consider the spin-inversion operator, $\mathcal{S}^x_T = \prod\limits_i S^x_i$ ( this form is valid for spin-1/2 only).

We can see that,

\begin{eqnarray}
[\mathcal{S}^x_T,H] &=& 0, \quad \textbf{Spin-inversion is a symmetric operation} \nonumber \\
\therefore \mathcal{S}^x_T|\psi\rangle &=& M_x|\psi\rangle, \quad M_x \in \{ \pm 1\}
\end{eqnarray}

## Construction of the basis

In the following we shall consider a spin-1/2 Heisenberg chain as a
practical example. In this model it is useful to represent the spins
pointing in the ‘up’ direction by a digit ‘1’, and the down-spins by a
‘0’. Following this rule, a state in the $S^{z}$ basis can be
represented by a sequence of ones and zeroes, i.e., a “binary word”.
Thus, two Néel configurations in the 4-site chain can be seen as
$$\begin{aligned}
\mid \uparrow \downarrow \uparrow \downarrow \rangle \equiv |1010\rangle , \\
\mid \downarrow \uparrow \downarrow \uparrow \rangle \equiv |0101\rangle .\end{aligned}$$
Once the up-spins have been placed the whole configuration has been
uniquely determined since the remaining sites can only be occupied by
down-spins. The resulting binary number can be easily converted into
integers $i\equiv
\sum_{l}b_{l}.2^{l}$, where the summation is over all the sites of the
lattice, and $b_{l}$ can be $1$ or $0$. For example:
$$\begin{array}{lllll}
2^{3} & 2^{2} & 2^{1} & 2^{0} &  \\
1 & 0 & 1 & 0 & \rightarrow 2^{1}+2^{3}=10, \\
0 & 1 & 0 & 1 & \rightarrow 2^{0}+2^{2}=5.
\end{array}$$

Using the above convention, we can construct the whole basis for the
given problem. However, we must consider the memory limitations of our
computer, introducing some symetries to make the problem more tractable.
The symetries are operations that commute with the Hamiltonian, allowing
us to divide the basis in subspaces with well defined quantum numbers.
By means of similarity transformations we can generate a sequence of
smaller matrices along the diagonal (i.e. the Hamiltonian matrix is
“block diagonal”), and each of them can be diagonalized independently.
For fermionic systems, the simplest symmetries are associated to the
conservation of the number of particles and the projection
$S_{total}^{z}$ of the total spin in the $z$ direction. In the spin-1/2
Heisenberg model, a matrix with $2^{N}\times 2^{N}$ elements can be
reduced to $2S+1$ smaller matrices, corresponding to the projections $%
m=(N_{\uparrow }-N_{\downarrow })/2=-S,-S+1,...,S-1,S$, with $S=N/2$.

The dimension of each of these subspaces is obtained from the
combinatory number $\left(
\begin{array}{c}
N \\
N_{\uparrow }
\end{array}
\right) = \frac{N!}{N_{\uparrow }!N_{\downarrow }!}$.

## Construction of the Hamiltonian matrix elements

1. Go through the loop of Hilbert space dimension.
2. Go through the inner loop involving the summation over bonds.
3. Calculate the diagonal matrix elements ($H_d$).
4. Calculate the off-diagonal matrix elements by connecting elements ($H_{od}$).

## Full Diagonalization implementation of 1D (chain) Heisenberg Model

```python
# -*- coding: utf-8 -*
from __future__ import absolute_import, division, print_function
import itertools
import numpy as np
import scipy as sp
from scipy import linalg
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation



def flip(l0, l1):         # Operation of flipping spins, lo, l1 = {-1,1}
    if (l0 == -1 and l1 == 1):
     return [[1, -1]]
    if (l0 == 1 and l1 == -1):
     return [[-1, 1]]

def basis_no(i,S):              # Convert spin configuration into appropriate index. Should be Bijective
    L=len(i)
    z=0
    for j in range(L):
        z += (S**j)*int((i[L-j-1]+1)/2)
    return z


def save(file,X,Y):
    np.savetxt(file, list(zip(X, Y)), fmt='%.12g')      # Save the file


def energy(T,W):       # Energy function, input is the eigenvalue matrix
    enrg=0
    Z=0
    for i in range(len(W)):
        Z       += np.exp(-(W[i])/T)           # Partition function
        enrg    += W[i]*np.exp(-W[i]/T)        # Energy ensemble average
    return (enrg/Z)  

def get_hamiltonian(L, J, s):
    S = int(round(2*s))+1    # Effective spin in the system
    nhil = S**L   # Hilbert space-dimension
    bonds = [[(i+j)%L for i in range(2)] for j in range(L)] # Array containing the interaction pairs
    H = np.zeros((nhil,nhil),dtype=np.float128) # Initialize the Hamiltonian Matrix
    for ii in itertools.product([-1,1], repeat=L): # Going through the Hilbert space loop
        diag_energy = 0     
        for bond in bonds: # Go through the bonds loop
            diag_energy+= -(J-J*(ii[bond[0]]*ii[bond[1]]))*s*s # Calculate the sum of diagonal matrix elements
            if ii[bond[0]] == ii[bond[1]]: # Parallel spins dont contribute to off-diaganal operator
               continue
            kk = flip(ii[bond[0]], ii[bond[1]]) # Flip and store the new config in kk
            for j in kk:
                jj = ii                
                jj = list(jj)
                jj[bond[0]]=j[0] 
                jj[bond[1]]=j[1] 
                jj = tuple(jj)

                H[basis_no(ii,S)][basis_no(jj,S)]=J*s
        H[basis_no(ii,S)][basis_no(ii,S)]=diag_energy
    return H 


def full_ED(L, J, s):
    X=[]
    Y=[]
    for T in np.arange(0.1,2.0,0.01):
        H = get_hamiltonian(L, J, s)
        W = sp.linalg.eigh(H, eigvals_only=True)
        e = energy(T,W)
        #print("Energy per spin at T =", T, "is,", e/L)
        X = np.append(X, np.array([T]))
        Y = np.append(Y, np.array([e/L]))
    save("enrg_"+str(J)+"_"+str(L)+".dat",list(X),list(Y)) 
```

# Computing in fixed $M_z$ subspace

```python
# -*- coding: utf-8 -*-
from __future__ import absolute_import, division, print_function
import numpy as np
import scipy as sp
from scipy import linalg
import matplotlib.pyplot as plt
 

def save(file,*args):
    np.savetxt(file, list(zip(*args)), fmt='%.12g')

def IBITS(n,i):
    return ((n >> i) & 1)

def Spin(x,n):
  "Get the n-th bit of the number x"
  return 2*(x & (1 << n) and 1 or 0)-1

def mag_basis(L, Nup):
    dim = 0
    nhil = 2**L
    basis = []
    for state in range(nhil):
        n_ones = 0
        for bit in range(L):
            n_ones += IBITS(state,bit)
        if(n_ones == Nup):
            basis.append(state)
            dim += 1
    return basis, dim

def full_basis(L):
    dim = 0
    nhil = 2**L
    basis = []
    for state in range(nhil):
        basis.append(state)
        dim += 1
    return basis, dim

def energy(T,W):
    enrg=0
    Z=0
    for i in range(len(W)):
        Z       += np.exp(-(W[i])/T)
        enrg    += W[i]*np.exp(-W[i]/T)
    return (enrg/Z)

def check_state(ii, L):  
    n_ones  = 0
    for bit in range(L):
         n_ones += IBITS(ii,bit)
    return n_ones

def get_mag_subspace_hamiltonian(L, J, s, Nup):
    basis, nred = mag_basis(L, Nup)
    S = int(round(2*s))+1    
    nhil = S**L
    if L == 2:
        bonds = [[0,1]]
    else:    
        bonds = [[(i+j)%L for i in range(2)] for j in range(L)]
    H = np.zeros((nred,nred),dtype=np.float64)
    for i in range(nred):
        ii = basis[i]
        diag_energy = 0     
        for bond in bonds:
            s1 = Spin(ii,bond[0])
            s2 = Spin(ii,bond[1])
            diag_energy += J*s1*s2*s*s
            if s1 == s2: # valid for spin-1/2 only
                continue
            mask = (1 << bond[0]) | (1 << bond[1])
            jj = (ii ^ mask)
            #print(ii, jj)
            j = basis.index(jj)
            H[i][j]=J/2
        H[i][i]=diag_energy                     
    return H


def get_hamiltonian(L, J, s, Nup):
    basis, nhil = full_basis(L)
    S = int(round(2*s))+1    
    nhil = S**L
    if L == 2:
        bonds = [[0,1]]
    else:    
        bonds = [[(i+j)%L for i in range(2)] for j in range(L)]
    H = np.zeros((nhil,nhil),dtype=np.float64)
    for i in range(nhil):
            ii = basis[i]
            diag_energy = 0     
            for bond in bonds:
                s1 = Spin(ii,bond[0])
                s2 = Spin(ii,bond[1])
                diag_energy += J*s1*s2*s*s
                if s1 == s2: # valid for spin-1/2 only
                    continue
                mask = (1 << bond[0]) | (1 << bond[1])
                jj = (ii ^ mask)
                #print(ii, jj)
                j = basis.index(jj)
                #print(ii,jj)
                H[i][j]=J*s
            H[i][i]=diag_energy
    return H

def mag_subspace_ED(L, J, s, Nup):
    X=[]
    Y=[]
    Y1=[]
    Y2=[]
    H = get_mag_subspace_hamiltonian(L, J, s, Nup)
    W = sp.linalg.eigh(H, eigvals_only=True) 
    H = get_hamiltonian(L, J, s, Nup)
    W2 = sp.linalg.eigh(H, eigvals_only=True)  
    print(W)
    for T in np.arange(0.1,3.0,0.01):
        e = energy(T,W)
        e2 = energy(T,W2)
        X = np.append(X, np.array([T]))
        Y = np.append(Y, np.array([e/L]))
        Y2 = np.append(Y2, np.array([e2/L]))
    save("mag_enrg_"+str(J)+"_"+str(L)+".dat",list(X),list(Y),list(Y2)) 
```

```python
if __name__ == '__main__':
    L = 6
    J = 1.0
    s = 1/2
    Nup = L/2
    plt.rcParams['font.size'] = '14'        
    for i in range(1):
        mag_subspace_ED(L, J, s, Nup)
        X, Y, Y2 = np.genfromtxt("mag_enrg_"+str(J)+"_"+str(L)+".dat", unpack=True, dtype=np.float64)
        plt.plot(X, Y, label="Subspace")
        plt.plot(X, Y2, label="Full")
        L += 2
    plt.xlabel('T')
    plt.ylabel('E')
    plt.xticks(np.arange(0, max(X)+0.5, 0.5))
    plt.yticks(np.arange(-0.50, max(Y)+0.1, 0.1))
    plt.title('Energy per spin V Temperature.')
    plt.legend(title='L='+str(L-2))
    plt.show()
```

# Computing in fixed $M_z$ + $\mathcal{I}$ operator's basis:

```python
def inv_basis(mag_basis, L):
    basis = mag_basis
    flip_basis = []
    dim = 0
    nhil = int(len(mag_basis)/2)
    #print(nhil)
    for i in range(nhil):
        state = basis[i]
        n_ones  = 1
        for bit in range(L):
            mask = (1 << bit)
            state = (state ^ mask)
        idx = basis.index(state)
        flip_basis.append(basis[idx])
        basis.pop(idx)
        dim += 1
    flip_basis.sort()    
    return basis, flip_basis, dim

def save2(file,X,Y):
    np.savetxt(file, list(zip(X, Y)), fmt='%.12g')

def get_inv_subspace_hamiltonian(L, J, s, Nup, M):
    basis, nred = mag_basis(L, Nup)
    basis1, basis2, nred = inv_basis(basis, L)
    #print(basis1, basis2, nred)
    S = int(round(2*s))+1    
    nhil = S**L
    if L == 2:
        bonds = [[0,1]]
    else:    
        bonds = [[(i+j)%L for i in range(2)] for j in range(L)]
    H = np.zeros((nred,nred),dtype=np.float64)
    for i in range(nred):
        ii = basis1[i]
        diag_energy = 0     
        for bond in bonds:
            s1 = Spin(ii,bond[0])
            s2 = Spin(ii,bond[1])
            diag_energy += J*s1*s2*s*s
            if s1 == s2: # valid for spin-1/2 only
                continue
            mask = (1 << bond[0]) | (1 << bond[1]) #flip
            jj = (ii ^ mask)
            #print(ii, jj)
            # THIS IS CRUCIAL STEP (NOTE THIS IMPLEMENTATION ESSENTIALLY TAKES 
            # THE "FIRST" SITE (I.E. THE MOST SIGNIFICANT BINARY DIGIT) AS THE REFERENCE SITE
            if jj > basis1[-1]:
                j = nred - 1 - basis2.index(jj)
                H[i][j]=M*J/2
            else:
                j = basis1.index(jj)
                H[i][j]=J/2
            #print("j is ", i,j)    
        H[i][i]=diag_energy  
    #print(H)    
    return H

def inv_subspace_ED(L, J, s, Nup, M):
    X=[]
    Y=[]
    H = get_inv_subspace_hamiltonian(L, J, s, Nup, M)
    W = sp.linalg.eigh(H, eigvals_only=True)
    print(W)
    for T in np.arange(0.1,3.0,0.01):
        e = energy(T,W)
        X = np.append(X, np.array([T]))
        Y = np.append(Y, np.array([e/L]))
    save2("inv_enrg_"+str(J)+"_"+str(L)+".dat",list(X),list(Y)) 
```

```python
if __name__ == '__main__':
    L = 8
    J = 1.0
    s = 1/2
    Nup = L/2
    M = 1
    plt.rcParams['font.size'] = '14'        
    for i in range(1):
        inv_subspace_ED(L, J, s, Nup, M)
        X, Y, Y2 = np.genfromtxt("mag_enrg_"+str(J)+"_"+str(L)+".dat", unpack=True, dtype=np.float64)
        X, Y1 = np.genfromtxt("inv_enrg_"+str(J)+"_"+str(L)+".dat", unpack=True, dtype=np.float64)
        plt.plot(X, Y, label="$M_z$")
        plt.plot(X, Y1, label="$M_z$+I")
        L += 2
    plt.xlabel('T')
    plt.ylabel('E')
    plt.xticks(np.arange(0, max(X)+0.5, 0.5))
    plt.yticks(np.arange(-0.50, max(Y)+0.1, 0.1))
    plt.title('Energy per spin V Temperature.')
    plt.legend(title='L='+str(L-2))
    plt.show()
```
